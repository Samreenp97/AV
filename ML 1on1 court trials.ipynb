```python
court=court.drop(columns=['Sub_Group_Name'])
train=pd.get_dummies(court)
train.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
arr=ss.fit_transform(train)
trainer=pd.DataFrame(arr,columns=train.columns)
trainer.head()
```

```python
from sklearn.model_selection import train_test_split
x=trainer[['Year', 'PT_1_3_Years', 'PT_5_10_Years',
       'PT_6_12_Months', 'PT_Less_than_6_Months', 'PT_Over_10_Years',
       'PT_Total', 'Area_Name_Andaman & Nicobar Islands',
       'Area_Name_Andhra Pradesh', 'Area_Name_Arunachal Pradesh',
       'Area_Name_Assam', 'Area_Name_Bihar', 'Area_Name_Chandigarh',
       'Area_Name_Chhattisgarh', 'Area_Name_Dadra & Nagar Haveli',
       'Area_Name_Daman & Diu', 'Area_Name_Delhi', 'Area_Name_Goa',
       'Area_Name_Gujarat', 'Area_Name_Haryana', 'Area_Name_Himachal Pradesh',
       'Area_Name_Jammu & Kashmir', 'Area_Name_Jharkhand',
       'Area_Name_Karnataka', 'Area_Name_Kerala', 'Area_Name_Lakshadweep',
       'Area_Name_Madhya Pradesh', 'Area_Name_Maharashtra',
       'Area_Name_Manipur', 'Area_Name_Meghalaya', 'Area_Name_Mizoram',
       'Area_Name_Nagaland', 'Area_Name_Odisha', 'Area_Name_Puducherry',
       'Area_Name_Punjab', 'Area_Name_Rajasthan', 'Area_Name_Tamil Nadu',
       'Area_Name_Tripura', 'Area_Name_Uttar Pradesh', 'Area_Name_Uttarakhand',
       'Area_Name_West Bengal', 'Group_Name_PT1. District/Session Judge',
       'Group_Name_PT2. Additional Session Judge',
       'Group_Name_PT3. Chief Judicial Magistrate',
       'Group_Name_PT4. Judicial Magistrate (I)',
       'Group_Name_PT5. Judicial Magistrate (II)',
       'Group_Name_PT6. Special Judicial Magistrate',
       'Group_Name_PT7. Other courts (if any)',
       'Group_Name_PT8. Total of Courts']]
y=trainer.PT_5_10_Years
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.5,random_state=1234)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain,ytrain)
ypred_lr=lr.predict(xtest)
print(r2_score(ytest,ypred_lr))
```

```python
from sklearn.tree import DecisionTreeRegressor
dr= DecisionTreeRegressor()
dr.fit(xtrain,ytrain)
ypred_dr=dr.predict(xtest)
print(r2_score(ytest,ypred_dr))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(xtrain,ytrain)
ypred_rf=rf.predict(xtest)
print(r2_score(ytest,ypred_rf))
```

```python
from sklearn.neighbors import KNeighborsRegressor
kn=KNeighborsRegressor(n_neighbors=7)
kn.fit(xtrain,ytrain)
ypred_kn=kn.predict(xtest)
print(r2_score(ytest,ypred_kn))
```

```python
fd1=pd.read_csv('C:\\Users\\Iqraa\\Downloads\\29_Period_of_trials_by_courts.csv')
fd1.head()
```

```python
fd1=fd1.drop(columns=['Sub_Group_Name'])
fd1=fd1.dropna(axis=0)
fd1.isnull().sum()
```

```python
fd1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
ff=fd1
le=LabelEncoder()
ff.Group_Name=le.fit_transform(ff.Group_Name)
ff.head()
```

```python
ff.columns
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
x=ff[[ 'Year', 'Group_Name', 'PT_1_3_Years', 'PT_3_5_Years',
       'PT_5_10_Years', 'PT_6_12_Months', 'PT_Less_than_6_Months',
       'PT_Over_10_Years', 'PT_Total']]
y=ff.Area_Name
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=12365)
dt=DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
ypred_dt=dt.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(ytest,ypred_dt))
print(confusion_matrix(ytest,ypred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(xtrain,ytrain)
ypred_rf=rf.predict(xtest)
print(accuracy_score(ytest,ypred_rf))
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier()
kn.fit(xtrain,ytrain)
ypred_kn=kn.predict(xtest)
print(accuracy_score(ytest,ypred_kn))
```
